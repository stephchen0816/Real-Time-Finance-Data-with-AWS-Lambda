```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\k00lg\Downloads\project03\Athena\results.csv')
```

```python
Average Volatility Trend per Company
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=df,x="Date",y="Avg_Volatility",hue="Company")
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Average Volatility",fontsize=15)
plt.title("Average Volatility Trend per Company",fontsize=20)
```

```python
The Daily Highest Volatility Per Company
```

```python
sns.catplot(data=df, kind="bar",x="Company", y="Max_Volatility", hue="Date",height=20,legend=False)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel("Company",fontsize=25)
plt.ylabel("Max Volatility",fontsize=25)
plt.title("Daily Highest Volatility Per Company",fontsize=25)
plt.legend(fontsize=15)
```
